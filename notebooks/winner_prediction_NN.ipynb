```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from torch.autograd import Variable
```

```python
input_size = data.shape[1]-1
hidden_size = 64
output_size = 2
BATCH_SIZE=128
EPOCH=100
```

```python
class winning(nn.Module) :
    
    def __init__(self, input_size, hidden_size, output_size) :
        super(winning, self).__init__()
        
        # activation function
        self.act = nn.ReLU()
            
        # neural network layers
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, output_size)
        self.linear3 = nn.Linear(hidden_size, output_size)
        
    def forward(self, X) :
        out = self.act(self.linear1(X))
        out = self.act(self.linear2(out))
        
        return out
```

```python
def batch(tensor, batch_size):
    tensor_list = []
    length = tensor.shape[0]
    i = 0
    while True:
        if (i+1) * batch_size >= length:
            tensor_list.append(tensor[i * batch_size: length])
            return tensor_list
        tensor_list.append(tensor[i * batch_size: (i+1) * batch_size])
        i += 1
```

```python
class Estimator(object):

    def __init__(self, model):
        self.model = model

    def compile(self, optimizer, loss):
        self.optimizer = optimizer
        self.loss_f = loss

    def _fit(self, X_list, y_list):
        """
        train one epoch
        """
        loss_list = []
        acc_list = []
        for X, y in zip(X_list, y_list):
            X_v = Variable(torch.FloatTensor(X))
            y_v = Variable(torch.LongTensor(y.reshape(BATCH_SIZE,)), requires_grad=False)


            self.optimizer.zero_grad()
            y_pred = self.model(X_v)
            
            loss = sum([self.loss_f(y_pred, y_v) for i in range(len(y_v))])/len(y_v)
            #print(y_pred.shape, y_v.shape)

            loss.backward()
            self.optimizer.step()

            ## for log
            loss_list.append(loss.data[0])

        return sum(loss_list) / len(loss_list)

    def fit(self, X, y, batch_size=BATCH_SIZE, nb_epoch=EPOCH, validation_data=()) :

        X_list = batch(X, batch_size)
        y_list = batch(y, batch_size)
        train_losses=[]
        val_losses=[]

        for t in range(1, nb_epoch + 1):
            loss = self._fit(X_list, y_list)
            val_log = ''
            if validation_data:
                val_loss = self.evaluate(validation_data[0], validation_data[1], batch_size)
                val_log = "- val_loss: %06.4f" % (val_loss)
            print("Epoch %s/%s loss: %06.4f  %s" % (t, nb_epoch, loss, val_log))
            train_losses.append(loss)
            val_losses.append(val_loss)

        return train_losses, val_losses


    def evaluate(self, X, y, batch_size=BATCH_SIZE):
        y_pred = self.predict(X)

        y_v = Variable(torch.from_numpy(y).float(), requires_grad=False)
        loss = sum([self.loss_f(y_pred[i], y_v[i]) for i in range(len(y_v))])/len(y_v)

        return loss.data[0]

    def predict(self, X):
        
        X = Variable(torch.FloatTensor(X))
        y_pred = self.model(X)
        
        return y_pred
```

```python
model = winning(input_size, hidden_size, output_size)

clf = Estimator(model)
clf.compile(optimizer=torch.optim.Adam(model.parameters(), lr=1e-4),
            loss=nn.CrossEntropyLoss())

train_losses, val_losses = clf.fit(np.array(X_train), np.array(y_train), batch_size=BATCH_SIZE, nb_epoch=EPOCH,
        validation_data=(np.array(X_test), np.array(y_test)))

score= clf.evaluate(np.array(X_test), np.array(y_test))
```
