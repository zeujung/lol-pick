```python
import pandas as pd
aa = pd.read_pickle('data2.pkl')
print(len(aa))

aa = [a for a in aa if a!=[]]  # remove empty history
df = pd.DataFrame(aa)
df = df.loc[df[20].isin([True, False])]
df.shape
```

```python
df.rename(columns=dict(zip(df.columns[[0,1,2,3,4]], ['blue_team_pick']*5)), inplace=True)
df.rename(columns=dict(zip(df.columns[[5,6,7,8,9]], ['blue_team_ban']*5)), inplace=True)
df.rename(columns=dict(zip(df.columns[[10,11,12,13,14]], ['red_team_pick']*5)), inplace=True)
df.rename(columns=dict(zip(df.columns[[15,16,17,18,19]], ['red_team_ban']*5)), inplace=True)
df.rename(columns={20: 'blue_team_win'}, inplace=True)
```

- First replace the id of the banned champs with champion name

```python
from cassiopeia import Champion, Champions

champions = Champions(region="KR")
id_to_name = {}
for champion in champions:
    id_to_name[champion.id] = champion.name
    #print(champion.name, champion.id)
```

```python
id_to_name[1]
```

```python
df.iloc[:,:-1]=df.iloc[:,:-1].replace(id_to_name)
```
